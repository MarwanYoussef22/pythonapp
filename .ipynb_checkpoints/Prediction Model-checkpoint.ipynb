```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report

# Load your dataset
df = pd.read_csv('troop_movements.csv')  # Replace 'character_data.csv' with your dataset file
print(df)
```

```python
# Encode categorical variables (homeworld and unit type)
#le = LabelEncoder()
#df['homeworld_encoded'] = le.fit_transform(df['homeworld'])
#df['unit_type_encoded'] = le.fit_transform(df['unit_type'])

# Define features (X) and target (y)
X = df[['homeworld', 'unit_type']]
Y = df['empire_or_resistance']  # Target variable

print(X)
print(Y)
```

```python
# Encode categorical variables (homeworld and unit type)
#le = LabelEncoder()
#df['homeworld_encoded'] = le.fit_transform(df['homeworld'])
#df['unit_type_encoded'] = le.fit_transform(df['unit_type'])
X_encoded = pd.get_dummies(X)
print(X_encoded)
```

```python
# Split the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X_encoded, Y, test_size=0.2, random_state=42)
print(train_test_split)
```

```python
# Create and train the Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=42)
clf.fit(X_train, Y_train)
```

```python
# Make predictions on the test set
Y_pred = clf.predict(X_test)
```

```python
# Evaluate the model
accuracy = accuracy_score(Y_test, Y_pred)
report = classification_report(Y_test, Y_pred)

print("Accuracy:", accuracy)
print("Classification Report:\n", report)
```

```python
importances = clf.feature_importances_
 
 
# Create a DataFrame to hold the feature importances
feature_importances = pd.DataFrame({'Feature': X_encoded.columns, 'Importance': importances})

print(feature_importances)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.barplot(data = feature_importances, x = 'Feature', y = 'Importance')
plt.xticks(rotation=90)
plt.title= ('Feature Importances')
plt.show
```
